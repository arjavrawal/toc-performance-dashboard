```python
from pathlib import Path
import geopandas as gpd
import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
DATA_DIR = PROJECT_ROOT / "data"
PROCESSED_BOUNDARIES_DIR = DATA_DIR / "boundaries" / "processed"
BOUNDARIES_DIR = DATA_DIR / "boundaries"
TOC_DIR = BOUNDARIES_DIR / "tocs"
VILLAGES_DIR = BOUNDARIES_DIR / "villages"

toc_gpkg = PROCESSED_BOUNDARIES_DIR / "toc_clean.gpkg"
villages_gpkg = PROCESSED_BOUNDARIES_DIR / "village_clean.gpkg"
toc_csv = TOC_DIR / "tocs_demographics_cleaned.csv"
villages_csv = VILLAGES_DIR / "villages_demographics_cleaned.csv"

toc_gpkg, villages_gpkg, toc_csv, villages_csv
```

```python
toc = gpd.read_file(toc_gpkg)
villages = gpd.read_file(villages_gpkg)
toc_df = pd.read_csv(toc_csv)
villages_df = pd.read_csv(villages_csv)

print("TOC GPKG:", len(toc))
print("TOC CSV:", len(toc_df))
print("Villages GPKG:", len(villages))
print("Villages CSV:", len(villages_df))
```

```python
print(toc.columns)
print(toc_df.columns)
```

```python
JOIN_FIELD_TOC = "APPLICABIL"
JOIN_FIELD_TOC_DF = "APPLICABIL"

JOIN_FIELD_TOC, JOIN_FIELD_TOC_DF
```

```python
toc_merge = toc.merge(
    toc_df,
    how="left",
    left_on=JOIN_FIELD_TOC,
    right_on=JOIN_FIELD_TOC_DF,
)

print("Rows after join (must match tract count):", len(toc_merge))
toc_merge.head()
```

```python
toc_merge_path = TOC_DIR / "toc_demographics.gpkg"
toc_merge.to_file(toc_merge_path, driver="GPKG")

print("Saved TOC+metrics layer to:", toc_merge_path)
```

```python
print(villages.columns)
print(villages_df.columns)
```

```python
JOIN_FIELD_VILLAGES = "NAME"
JOIN_FIELD_VILLAGES_DF = "NAME"

JOIN_FIELD_VILLAGES, JOIN_FIELD_VILLAGES_DF
```

```python
village_merge = villages.merge(
    villages_df,
    how="left",
    left_on=JOIN_FIELD_VILLAGES,
    right_on=JOIN_FIELD_VILLAGES_DF
)

print("Rows after join (must match tract count):", len(village_merge))
village_merge.head()
```

```python
village_merge_path = VILLAGES_DIR / "village_demographics.gpkg"
village_merge.to_file(village_merge_path, driver="GPKG")

print("Saved village+metrics layer to:", village_merge_path)
```
