```python
from pathlib import Path
import geopandas as gpd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
DATA_DIR = PROJECT_ROOT / "data"
BOUNDARIES_DIR = DATA_DIR / "boundaries"

BOUNDARIES_DIR
```

```python
raw_village_path = BOUNDARIES_DIR / "villages" / "Villages.shp"

village_clean_path = BOUNDARIES_DIR / "processed" / "village_clean.gpkg"

raw_village_path, village_clean_path
```

```python
village = gpd.read_file(raw_village_path)

print("Village features:", len(village), "| CRS:", village.crs)

village.head()
```

```python
PROJECT_CRS = "EPSG:2868"

def ensure_project_crs(gdf, name="layer"):
    if gdf.crs is None:
        print(f"{name}: CRS is None – set manually before reprojecting.")
        gdf = gdf.set_crs("EPSG:2868")
        return gdf
    
    if PROJECT_CRS is not None and gdf.crs != PROJECT_CRS:
        print(f"{name}: Reprojecting from {gdf.crs} to {PROJECT_CRS}")
        return gdf.to_crs(PROJECT_CRS)
    else:
        print(f"{name}: No reprojection performed.")
        return gdf

village = ensure_project_crs(village, "Village")
```

```python
village.columns
```

```python
VILLAGE_ID_FIELD = "OBJECTID"
VILLAGE_NAME_FIELD = "NAME"

village_clean = village[[VILLAGE_ID_FIELD, VILLAGE_NAME_FIELD, "geometry"]].copy()

village_clean.head()
```

```python
village_clean.to_file(village_clean_path, driver="GPKG")

print("Saved cleaned file:", village_clean_path)
```
