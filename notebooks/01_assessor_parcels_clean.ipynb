# Assessor & Parcel Data - Cleaning and Standardization

This notebook prepares the **Maricopa County assessor and parcel geometry data** 
for use in the TOC Performance Dashboard.

The aim is to have:
- A cleaned **parcel-level GeoDataFrame** with:
  - consistent parcel ID field,
  - basic land use / property class attributes,
  - improvement and land value fields,
  - lot area (square feet and acres),
- A spatial layer clipped (or at least limited) to **Maricopa County**,
- Saved outputs to `data/parcels/processed/` for downstream notebooks.

This becomes the base layer for:
- calculating taxable value per acre,
- aggregating to TOC station areas and village boundaries,
- and linking parcels to job counts and zoning where possible.
